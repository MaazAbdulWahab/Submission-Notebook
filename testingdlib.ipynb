```python
import dlib
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
detector = dlib.cnn_face_detection_model_v1('dogHeadDetector.dat')
```

```python
img=cv2.imread('cute-dog.jpg')
print(type(img))
print(img.shape)
```

```python
plt.imshow(img)
```

```python
dets = detector(img, upsample_num_times=1)
```

```python
print(dets)
```

```python
left=dets[0].rect.left()
top=dets[0].rect.top()
right=dets[0].rect.right()
bottom=dets[0].rect.bottom()
```

```python
img=cv2.rectangle(img, (left,top), (right,bottom), (0,255,0), 5)
```

```python
#cv2.imshow('dog',img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(img)
```

```python
plt.imshow(img[top:bottom,left:right])
```
